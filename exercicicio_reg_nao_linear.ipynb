```python
#Importando libs 
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn import metrics 
```

```python
#Importando dataset de casas 
df = pd.read_excel('house (1).xlsx')
df.head()
```

Análise Exploratória

```python
df.info()

#Encontramos valores nulos nas colunas 'LotFrontage' e 'MasVnrArea'
#Também encontramos categóricas
```

```python
#Vendo valores nulos das respectivas colunas 
df[df['LotFrontage'].isnull()]
df[df['MasVnrArea'].isnull()]

# Dropando valores os valores nulos do df
df = df.dropna()
```

```python
#Transformando colunas categoricas 
df = pd.get_dummies(df)
df
```

```python
#Análise descritiva 
df.describe()
```

Definindo variáveil Label e as variáveis features (x)

```python
labels = np.array(df['SalePrice']) #criando um array da minha target 'SalePrice'
features = df.drop('SalePrice', axis=1) #features = todo o restante do meu df, tirando o 'SalePrice'

features_list = list(df.columns) #criando uma lista das colunas do dataset 

df = np.array(features) #definindo meu dataset com um array 
```

Variáveis de treino e teste

```python
#Definindo minhas variáveis de treino e de teste 
train_features, test_features, train_labels, test_labels = train_test_split(features,labels, test_size= 0.25,random_state=42) #restando para treino
```

Treino com RandomForestRegressor

```python
#Definindo Rf 
rf = RandomForestRegressor(n_estimators=100, random_state=42)

#Treinando variável 
rf.fit(train_features,train_labels)
```

```python
#Fazendo a predição 
predictions_rf = rf.predict(test_features)

#Calculando o erro médio absoluto (abs)
errors_rf = abs(predictions_rf-test_labels)

#Determinando o coeficiente de determinação R² 
r_sq = rf.score(features,labels)
print(r_sq)

#MAE
print(metrics.mean_absolute_error(test_labels,predictions_rf,))

#MSE 
print(metrics.mean_squared_error(test_labels,predictions_rf))

#RMSE
print(np.sqrt(metrics.mean_squared_error(test_labels,predictions_rf)))


```

AdaBoostRegressor 

```python
#Definindo o Ada
ada = AdaBoostRegressor(n_estimators=100) #como esse modelo não trabalha de forma aleatoria, não definimos o random state.

#Treinando o modelo 
ada.fit(train_features,train_labels)

#Fazendo o predição 
ada_pred = ada.predict(test_features)
```

```python
#Calculando o erro médio absoluto (abs)
errors_ada = abs(ada_pred - test_labels)

#R² 
r_sq = ada.score(features,labels)
print(r_sq)

#MAE 
print(metrics.mean_absolute_error(test_labels,ada_pred))

#MSE
print(metrics.mean_squared_error(test_labels,ada_pred))

#RMSE 
print(np.sqrt(metrics.mean_squared_error(test_labels,ada_pred)))
```

GradientBoostingRegressor 

```python
#Definindo gb 
gb = GradientBoostingRegressor(n_estimators=100)

#Treinando o modelo 
gb.fit(train_features,train_labels)

#Fazendo o predição 
gb_pred = gb.predict(test_features)
```

```python
#Calculando o erro médio absoluto (abs)
errors_gb = abs(gb_pred - test_labels)

#R²
r_sq = gb.score(features,labels)
print(r_sq)

#MAE  
print(metrics.mean_absolute_error(test_labels,gb_pred))

#MSE  
print(metrics.mean_squared_error(test_labels,gb_pred))

#RMSE 
print(np.sqrt(metrics.mean_squared_error(test_labels,gb_pred)))
```

Identificando as features do meu melhor modelo ( Gradient Boosting Regressor )

```python
#Fazendo a identificação das features usadas do modelo 
importances = list(gb.feature_importances_)

#Listando as tuplas com suas variaveis e as importancias 
feature_importances = [(features,round(importances, 2)) for features, importances in zip(features_list,importances)]

#Ordenando por valores mais importantes aos menores 
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

#Printando as variaveis de acordo com a importancia 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
