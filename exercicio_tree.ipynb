```python
#Importando libs 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import tree
```

```python
#Carregando dataset 
df = pd.read_excel('vendas.xlsx')
df.head()
```

Análise Exploratória 

```python
df.info()
df.shape
```

Convertendo colunas categoricas em numéricas 

```python
#Importando lib LabelEncoder
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()

#Transformando minhas variáveis categoricas em numéricas 
df['test'] = le.fit_transform(df['test'])
df['cliente_comprou'] = le.fit_transform(df['cliente_comprou']) #cliente comprou = 1, não comprou = 0
```

```python
#Definindo x e y 
x = df.iloc[:,:-1]
y = df['cliente_comprou']
```

Decision Tree

```python
#Com a lib já importada, definimos ela 
decision_tree = tree.DecisionTreeClassifier()

#Aplicando 
model = decision_tree.fit(x,y)
model
```

Criando a visulização da arvore 

```python
fig = plt.figure(figsize=(20,10))

fig = tree.plot_tree(model,
                     feature_names= x.columns,
                     class_names= ['Sim','Não'],
                     filled= True  #preenche os nós com cores 
 )
```
